```python
import datetime
```

```python
#---- Obtain class type
def getClassType(classTypeString):
    return classTypeString

#---- Obtain group
def getGroup(groupString):
    return groupString

#---- day
def getDay(dayString):
    # Convert date into integer
    dayDict = {'MON' : 0, 'TUE' : 1, 'WED' : 2, 'THU' : 3, 'FRI' : 4, 'SAT' : 5}
    return dayDict[dayString]

#---- time
def getDuration(timeString):
    timestamp1 = datetime.datetime.strptime(timeString.split('-')[0],"%H%M")
    timestamp2 = datetime.datetime.strptime(timeString.split('-')[1],"%H%M")
    return (timestamp2 - timestamp1).total_seconds()

#---- venue
def getVenue(venueString):
    return venueString
```

```python
def getStartDate(firstDayOfSchool = datetime.date(2021,8,9)): # get based on fixed value first
    return firstDayOfSchool - datetime.timedelta(days=firstDayOfSchool.weekday())

def getModuleStartWeek(remark, day, startDate):
    # find starting week
    return int(remark[remark.find('Wk')+2])

def getRecurrences(remark):
    remark = remark[remark.find('Wk')+2:]
    # find if there are any commas
    if ',' in remark: # bi-weekly schedule
        return len(remark.split(','))
    else:
        a,b = remark.split('-')
        return b-a+1
def isBiWeekly(remark):
    remark = remark[remark.find('Wk')+2:]
    # find if there are any commas
    if ',' in remark: # bi-weekly schedule
        return True
    return False
```

```python
import datetime
import pytz
import numpy as np

cal = Calendar()

#---- Find out what is the first day of school
#---- Find out how many teaching weeks are there
#---- Find out when is recess week

startDate = datetime.date.today()

#---- Iterate through the rows ----
for i in range(len(df.index)):
    #---- If row is not Nan, move to next module
    if pd.isnull(df.iloc[i]['Course'])or i==0:
        #---- Obtain class type
        title = getClassType(df.iloc[i]['Title'])
        course = df.iloc[i]['Course']
    #---- If row is Nan, still on the same module ----
    #---- Obtain group
    group = df.iloc[i]['Group']
    indexNumber = df.iloc[i]['IndexNumber']
    classType = getClassType(df.iloc[i]['ClassType'])

    #---- day, implement on start week
    dayInt = getDay(df.iloc[i]['Day'])
    startDay = getModuleStartWeek(df.iloc[i]['Remark'], dayInt, startDate)
    
    #---- time, implement bi or weekly
    time = getDuration(df.iloc[i]['Time']) # time in seconds
    biWeekly = isBiWeekly(df.iloc[i]['Remark'])
    #---- venue
    venue = df.iloc[i]['Venue']
    # obtain the end date
    endDate = startDate
    # ---- Add to event
    event = Event()
    event.add('summary', "{} {} {}".format(course, classType, title))
    event.add('description', indexNumber)
    event.add('dtstart', startDate)
    event.add('duration', datetime.timedelta(seconds=time))
    event.add('dtstamp', datetime.datetime.today())
    event.add('location', venue)
    event.add('rrule', {'freq' : 'weekly', 'until' : endDate})
    cal.add_component(event)
    if i == 5:
        break

f = open('{}.ics'.format('Test'), 'wb')
f.write(cal.to_ical())
f.close()
```
